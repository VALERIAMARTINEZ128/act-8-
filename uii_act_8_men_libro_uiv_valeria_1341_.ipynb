Una empresa tiene el registro de las horas que trabaja diariamente un

* (4.11):
Una persona se encuentra en el kilómetro 70 de la carretera Aguascalientes

viaja en dirección a Zacatecas, mientras que la segunda se dirige a Aguascalientes, a la misma velocidad

```python
import math
# Menu de problemas
def VerMenu():
    print("\n--- MENÚ DE ALGORITMOS (CAPÍTULO 4) ---")
    print("1. Suma de 10 cantidades (Ciclo While)")
    print("2. Suma de 10 cantidades (Ciclo Repeat/Until)")
    print("5. Estatura promedio (Grupo desconocido - Repeat)")
    print("10. Total de horas y sueldo semanal (6 días)")
    print("11. Punto de encuentro de dos vehículos (KM 70 y 150)")
    print("12. Salir")
    print("----------------------------------")
    opcion = int(input("Ingrese una opción: "))
    return opcion

# (4.1) Suma de 10 cantidades con While
def SumaDiezWhile():
    print("--- Suma de 10 cantidades (While) ---")
    contador = 1
    suma = 0
    while contador <= 10:
        valor = float(input(f"Ingrese la cantidad {contador}: "))
        suma = suma + valor
        contador = contador + 1
    print(f"La suma total es: {suma}")
    print("----------------------------------")

# (4.2) Suma de 10 cantidades con Repeat (Simulado)
def SumaDiezRepeat():
    print("--- Suma de 10 cantidades (Repeat) ---")
    contador = 1
    suma = 0
    while True: # Inicio del Repeat
        valor = float(input(f"Ingrese la cantidad {contador}: "))
        suma = suma + valor
        contador = contador + 1
        if contador > 10: # Condición Until
            break
    print(f"La suma total es: {suma}")
    print("----------------------------------")

# (4.5) Estatura promedio de grupo desconocido
def EstaturaPromedio():
    print("--- Promedio de Estaturas ---")
    suma_estaturas = 0
    cuenta_personas = 0
    while True:
        estatura = float(input("Ingrese estatura (o use 0 para terminar): "))
        if estatura == 0:
            break
        suma_estaturas += estatura
        cuenta_personas += 1

    if cuenta_personas > 0:
        promedio = suma_estaturas / cuenta_personas
        print(f"El promedio de estatura de las {cuenta_personas} personas es: {promedio:.2f}m")
    else:
        print("No se ingresaron estaturas.")
    print("----------------------------------")

# (4.10) Horas trabajadas y sueldo semanal
def SueldoSemanal():
    print("--- Registro de Horas (6 días) ---")
    total_horas = 0
    pago_por_hora = float(input("Ingrese el pago por hora: "))
    dia = 1
    while dia <= 6:
        horas = float(input(f"Horas trabajadas el día {dia}: "))
        total_horas += horas
        dia += 1

    sueldo = total_horas * pago_por_hora
    print(f"Total de horas en la semana: {total_horas}")
    print(f"Sueldo total a recibir: ${sueldo:.2f}")
    print("----------------------------------")

# (4.11) Encuentro de vehículos (KM 70 y 150)
def PuntoEncuentro():
    print("--- Punto de Encuentro ---")
    # Al ir a la misma velocidad, el punto de encuentro es el punto medio
    km_primera = 70
    km_segunda = 150

    # Lógica: Mientras no se crucen
    distancia = km_segunda - km_primera
    punto_encuentro = km_primera + (distancia / 2)

    print(f"Vehículo 1 en KM {km_primera}, Vehículo 2 en KM {km_segunda}")
    print(f"Se encontrarán en el kilómetro: {punto_encuentro}")
    print("----------------------------------")

# --- Ciclo Principal ---
while True:
    try:
        op = VerMenu()
        if op == 1:
            SumaDiezWhile()
        elif op == 2:
            SumaDiezRepeat()
        elif op == 3:
            EstaturaPromedio()
        elif op == 4:
            SueldoSemanal()
        elif op == 5:
            PuntoEncuentro()
        elif op == 6:
            print("Gracias por usar el programa, Keyla.")
            break
        else:
            print("Opción incorrecta.")
    except ValueError:
        print("Error: Ingrese un número válido.")

print("----------------------------------")
print4("TERMINADO POR VALERIA YARETZI ")
```
